```python
import pandas as pd
import numpy as np
# Load the data
data = pd.read_csv('/content/drive/MyDrive/Colab(AI)/Salary Prediction/Salary_Data.csv')
```

```python
# Define the features and target
features = data.loc[:3,'YearsExperience'].values
labels = data.loc[:3,'Salary'].values
features=features.reshape(4,-2)
labels=labels.reshape(4,-3)
features
#labels
```

```python
from sklearn.model_selection import train_test_split

# Split the data into training and test sets
features_train, features_test, labels_train, labels_test = train_test_split(features,labels, test_size=0.25, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
# Create the model
model = LinearRegression()
```

```python
# Fit the model using the training data
model.fit(features_train,labels_train)
```

```python
# Make predictions using the model
labels_pred = model.predict(features_test)
```

```python
print(labels_pred.shape)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Calculate the mean absolute error and mean squared error
mae = mean_absolute_error(features_test, labels_pred)
mse = mean_squared_error(features_test, labels_pred)

# Print the results
print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)
```

```python
model.score(features_test,labels_test)
```
